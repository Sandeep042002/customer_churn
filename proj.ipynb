```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

def one_hot_encode(df, cols):
    df = df.copy()
    encoder = OneHotEncoder(sparse_output=False, drop="first")  # Avoid multicollinearity
    encoded_df = pd.DataFrame(encoder.fit_transform(df[cols]))
    
    # Rename columns for clarity
    encoded_df.columns = encoder.get_feature_names_out(cols)
    
    # Drop original categorical columns & concat new ones
    df = df.drop(columns=cols).reset_index(drop=True)
    df = pd.concat([df, encoded_df], axis=1)

    return df
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV
import xgboost as xgb

 

# 1. Train basic models first
def train_basic_models(X_train, y_train, X_test, y_test):
    # Random Forest
    rf = RandomForestClassifier(random_state=42)
    rf.fit(X_train, y_train)
    rf_pred = rf.predict(X_test)
    
    # XGBoost
    xgb_model = xgb.XGBClassifier(random_state=42)
    xgb_model.fit(X_train, y_train)
    xgb_pred = xgb_model.predict(X_test)
    
    # Print results
    print("Basic Random Forest Accuracy:", accuracy_score(y_test, rf_pred))
    print("Basic XGBoost Accuracy:", accuracy_score(y_test, xgb_pred))
    
    return rf, xgb_model

# 2. Simple hyperparameter tuning
def tune_models(X_train, y_train):
    # Random Forest tuning - simple grid
    rf_params = {
        'n_estimators': [100, 200],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5]
    }
    
    rf = RandomForestClassifier(random_state=42)
    rf_grid = GridSearchCV(rf, rf_params, cv=3, n_jobs=-1)
    rf_grid.fit(X_train, y_train)
    
    print("Best Random Forest Parameters:", rf_grid.best_params_)
    
    # XGBoost tuning - simple grid
    xgb_params = {
        'n_estimators': [100, 200],
        'max_depth': [3, 6],
        'learning_rate': [0.1, 0.01]
    }
    
    xgb_model = xgb.XGBClassifier(random_state=42)
    xgb_grid = GridSearchCV(xgb_model, xgb_params, cv=3, n_jobs=-1)
    xgb_grid.fit(X_train, y_train)
    
    print("Best XGBoost Parameters:", xgb_grid.best_params_)
    
    return rf_grid.best_estimator_, xgb_grid.best_estimator_

# 3. Evaluate tuned models
def evaluate_models(rf_tuned, xgb_tuned, X_test, y_test):
    # Predictions
    rf_pred = rf_tuned.predict(X_test)
    xgb_pred = xgb_tuned.predict(X_test)
    

    print("\nTuned Random Forest Results:")
    print("Accuracy:", accuracy_score(y_test, rf_pred))
    print(classification_report(y_test, rf_pred))
    
    print("\nTuned XGBoost Results:")
    print("Accuracy:", accuracy_score(y_test, xgb_pred))
    print(classification_report(y_test, xgb_pred))
    
    # Feature importance
    if hasattr(rf_tuned, 'feature_importances_'):
        rf_importance = pd.DataFrame({
            'Feature': X_test.columns,
            'Importance': rf_tuned.feature_importances_
        }).sort_values('Importance', ascending=False)
        
        print("\nRandom Forest Top 10 Features:")
        print(rf_importance.head(10))
    
    if hasattr(xgb_tuned, 'feature_importances_'):
        xgb_importance = pd.DataFrame({
            'Feature': X_test.columns,
            'Importance': xgb_tuned.feature_importances_
        }).sort_values('Importance', ascending=False)
        
        print("\nXGBoost Top 10 Features:")
        print(xgb_importance.head(10))

# Run the whole process
def run_models(X_train, X_test, y_train, y_test):
    # Step 1: Train basic models
    print("Training basic models...")
    rf_basic, xgb_basic = train_basic_models(X_train, y_train, X_test, y_test)
    

    print("\nTuning models...")
    rf_tuned, xgb_tuned = tune_models(X_train, y_train)
    print("\nEvaluating tuned models...")
    evaluate_models(rf_tuned, xgb_tuned, X_test, y_test)
    
    return rf_tuned, xgb_tuned

rf_model, xgb_model = run_models(X_train, X_test, y_train, y_test)
```

```python
import os
import joblib

def save_model(model, filename, folder="models_path"):
    # Create folder if it doesn't exist
    if not os.path.exists(folder):
        os.makedirs(folder)
    
    model_path = os.path.join(folder, filename)
    joblib.dump(model, model_path)
    
    print(f"Model saved at {model_path}")


save_model(rf_model, "best_random_forest.pkl")
save_model(xgb_model, "best_xgboost.pkl")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from torch.utils.data import TensorDataset, DataLoader
```

```python
X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).unsqueeze(1)  # Add an extra dimension for the target
X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).unsqueeze(1)

# Create dataset loaders
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self, input_dim):
        super(NeuralNetwork, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        x = self.sigmoid(x)
        return x

input_dim = X_train_tensor.shape[1]
model = NeuralNetwork(input_dim)
```

```python
num_samples = len(y_train_tensor)
num_positive = torch.sum(y_train_tensor).item()
num_negative = num_samples - num_positive
class_weights = torch.tensor([num_samples / (2 * num_negative), num_samples / (2 * num_positive)]).float()
```

```python
criterion = nn.BCEWithLogitsLoss(pos_weight=class_weights[1])
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
num_epochs = 100
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}')
```

```python
model.eval()
with torch.no_grad():
    y_pred_prob = model(X_test_tensor).numpy()
    y_pred = (y_pred_prob > 0.5).astype(int)
```

```python
def load_model(model_path, input_dim):
    # Create a new instance of the model
    model = NeuralNetwork(input_dim)
    
    # Load the state dictionary
    model.load_state_dict(torch.load(model_path, weights_only=True))
    model.eval()
    
    return model
```
